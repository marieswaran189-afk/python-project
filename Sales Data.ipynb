```python
quantity = np.array([10,5,8,15,12])
unit_price=np.array([50,30,20,60,25])
profit=np.array([80,40,25,120,70])
```

```python
total_sales = quantity * unit_price

total_revenue = np.sum(total_sales)
average_sales = np.mean(total_sales)

print("Total Revenue:", total_revenue)
print("Average Sales:", average_sales)
```

```python
print("First 3 sales records:", total_sales[:3])
print("Last 2 sales records:", total_sales[-2:])
print("Second record sale:", total_sales[1])
```

```python
mean_sales = np.mean(total_sales)
median_sales = np.median(total_sales)
variance_sales = np.var(total_sales)
std_sales = np.std(total_sales)

print("Mean:", mean_sales)
print("Median:", median_sales)
print("Variance:", variance_sales)
print("Standard Deviation:", std_sales)
```

```python
revenue = quantity * unit_price
print("Revenue using broadcasting:", revenue)
```
